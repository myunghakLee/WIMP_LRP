```python
upper = sorted(fdes)[-(len(fdes) // 10)]
lower = sorted(fdes)[(len(fdes) // 10)]
upper, lower
```

```python

def get_ranking(agent, social, length):
    actors_feature = np.concatenate((np.expand_dims(agent, axis=0), social))
    return np.argsort(np.sum(actors_feature, axis=(1,2))[:length])

def draw_scatter(d, n):
    plt.scatter(n, d[0], color = 'blue', s = 60)
    plt.scatter([n for _ in range(len(d)-1)], d[1:], color = 'red', s = 60)

draw = False

def calc_metric(data, draw = False):
    if draw:
        plt.figure(figsize = (5, 15))

    total_score1 = 0
    total_score2 = 0
    total_score3 = 0
    for d in data:
        num_agent = max(np.nonzero(d['adjacency_lrp'][0])[0])+1
        agent_ranking = get_ranking(d['agent_features_lrp'], d['social_features_lrp'], num_agent)
        center_line_ranking = get_ranking(d['agent_oracle_centerline_lrp'], d['social_oracle_centerline_lrp'], num_agent)
        gan_feature_ranking = get_ranking(d['gan_features_lrp'][0], d['gan_features_lrp'][1:], num_agent)
        adjacency_lrp = get_ranking(np.argsort(d['adjacency_lrp'][0], num_agent))
        graph_output_s_lrp = get_ranking(d['graph_output_s_lrp'][0], d['graph_output_s_lrp'][1:], num_agent)
        
        print(adjacency_lrp)
        if draw:
            draw_scatter(agent_ranking, 0) # centerline
            draw_scatter(gan_feature_ranking, 1)
            draw_scatter(graph_output_s_lrp, 2) # adjacency
            pl = np.concatenate(([agent_ranking], [gan_feature_ranking], [graph_output_s_lrp]), axis=0)
            plt.plot(pl, color = 'gray')
        
#         agent_ranking = (agent_ranking + center_line_ranking) / 2
        agent_ranking = center_line_ranking
        
        total_score1 += np.sum(abs(agent_ranking - gan_feature_ranking) )/num_agent
        total_score2 += np.sum(abs(gan_feature_ranking - graph_output_s_lrp) )/num_agent
        total_score3 += np.sum(abs(agent_ranking - graph_output_s_lrp) )/num_agent

    if draw:
        plt.show()
#     print(total_score / len(data))
    print((total_score1/len(data)).round(5), (total_score2/len(data)).round(5), (total_score3/len(data)).round(5))
#     return total_score / len(data)

# draw_scatter(agent_ranking, 1)
```
